### Converting all quarters in 'Date' into date format

```python
# CP_df
CP_df['Date'] = pd.to_datetime(CP_df['Date'])
CP_df['Date'] = CP_df['Date'].dt.to_period('Q')

CP_df['Date'] = CP_df['Date'].dt.to_timestamp()

CP_df
```

```python
# expen_df
expen_df['Date'] = pd.to_datetime(
     expen_df['Date']
    .str.replace('Q1', '01').str.replace('Q2', '04')
                   .str.replace('Q3', '07').str.replace('Q4', '10')
)
```

### Writing the Model

```python
CP_df['Category'].unique()
```

```python
expen_df['Category'].unique()
```

```python
# List of categories to select
selected_categories = ['Dairy Products & Eggs', 'Vegetables & Fruit', 'Fish & Seafood']

# Filter CPI/PPI dataframe for the selected categories
filtered_CP = CP_df[CP_df['Category'].isin(selected_categories)]
filtered_CP
```

```python
filtered_CP
```

```python
grouped_CP = filtered_CP.groupby('Category').mean()
print(grouped_CP)
```

```python
filtered_CP_aggregated = filtered_CP.groupby(['Date', 'Category'])[['CPI_Value', 'PPI_Value']].mean().reset_index()
```

```python
# Select specific columns 'CPI_Value' and 'PPI_Value' and group by 'Date' and 'Category', then calculate the mean
filtered_CP_aggregated = filtered_CP[['Date', 'Category', 'CPI_Value', 'PPI_Value']].groupby(['Date', 'Category']).mean().reset_index()

# Display the aggregated data
print(filtered_CP_aggregated)
```

```python
# Filter rows for the date range 20192021
filtered_CP_aggregated = filtered_CP_aggregated[60:96]

filtered_CP_aggregated
```

```python
filtered_CP_aggregated
```

```python
category_mapping = {
    "Dairy Products & Eggs": "Hen Shell Eggs",
    "Fish & Seafood": "Vegetables",
    "Vegetables & Fruit": "Seafood",
}

filtered_CP_aggregated['Category'] = filtered_CP_aggregated['Category'].map(category_mapping)
```

```python
linear_df = pd.merge(
    expen_df,
    filtered_CP_aggregated,
    on=['Date', 'Category'],  # Merge on both date and category
    how='inner'              # Use inner join to keep only matching rows
)
```

```python
linear_df = linear_df.sort_values(by=['Category', 'Date'])
```

```python
cat_dummies = pd.get_dummies(linear_df.Category,drop_first=True)
```

```python
# combining this with our linear regression dataframe
linear_df = pd.concat([linear_df,cat_dummies],axis=1)
```

### Linear Regression Model

```python
linear_df.columns
```

```python
feature_cols = ['CPI_Value','PPI_Value', 'Seafood', 'Vegetables']
X = linear_df[feature_cols]
y = linear_df['Expenditure (S$ Million)']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both train and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize and fit the Linear Regression model
linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
```

```python
# Print the coefficients.
print(linreg.intercept_)
print(linreg.coef_)
```

```python
# y value is 143.41881698071427 when all x variables = 0
```

```python
correlations = linear_df[feature_cols].corr()
print(correlations)
print(sns.heatmap(correlations,annot=True))
```

```python
# Analyze Residuals
# Compute residuals
y_pred = linreg.predict(X_test)
residuals = y_test - y_pred

# Plot residuals
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted values
plt.figure(figsize=(8, 6))

# Scatter plot of actual vs predicted
plt.scatter(y_test, y_pred, alpha=0.7, color='blue', label='Predicted')

# Reference line (Perfect Prediction Line)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()],
         color='red', linestyle='--', linewidth=2, label='Line of Best Fit (Actual)')

# Add title, labels, and legend
plt.title("Predicted vs Actual Values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.show()

import numpy as np

# Index for each data point
index = np.arange(len(y_test))

# Plot actual and predicted values
plt.figure(figsize=(12, 6))
plt.scatter(index, y_test, color='blue', label='Actual Values', alpha=0.7)
plt.scatter(index, y_pred, color='orange', label='Predicted Values', alpha=0.7)

# Connect actual and predicted values with a line to show errors
for i in range(len(y_test)):
    plt.plot([index[i], index[i]], [y_test.iloc[i], y_pred[i]], color='gray', alpha=0.5)

# Add labels, title, and legend
plt.title("Actual vs Predicted Values with Errors")
plt.xlabel("Data Point Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred,squared = False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (MSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R): {r2}")
print(f"Accuracy via RMSE(*1.96): {rmse*1.96}%")
print(f"Mean Absolute Percentage Error (MAPE): {mape*100}%")
```
